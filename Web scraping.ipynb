```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.marketsandmarkets.com/telecom-and-IT-market-research-113.html'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id='publishedreports')
```

```python
print(results.prettify())
```

```python
## 
```

```python
python_jobs = results.find_all('a')

print(python_jobs)
```

## TITLE

```python
title = []
for i in python_jobs:
    title.append(i.get_text())
```

```python
title
```

## DESCRIPTION

```python
python_jobs1 = results.find_all('p')
descp = []
for i in python_jobs1:
    descp.append(i.get_text())
descp
```

## DATE

```python
python_jobs2 = results.find_all('td')
td = []
for i in python_jobs2:
    td.append(i.get_text())
td
```

```python
len(td)
```

```python
import re
regex = re.compile(r'[\n\xa0]')
```

```python
pub=[]

for n in range(200):
    ind = 1 + 4*n
    if ind <200:
        s = regex.sub(" ", td[ind])
        pub.append(s)
        
    else:
        break
pub    
```
